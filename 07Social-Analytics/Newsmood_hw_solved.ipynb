```python
# Importing Dependencies
import pandas as pd
import json
import tweepy
import matplotlib.pyplot as plt
import numpy as np 
#import and initialize Vader Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
# Import twitter API keys
from config import consumer_key, consumer_secret, access_token, access_token_secret
```

```python
# Twitter Credentials
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
# Running and appending the recent 100 tweets for BBC News
# Declaring the search term
target_term = "@BBC"
# Initializing the list
tweets_BBC = []
# Variable for holding sentiments
sentiments_BBC = []
# counter
counter = 1
# Get all tweets from home page
public_tweets = api.user_timeline(target_term, count=100, result_type="recent")
# loop through all tweets
for tweet in public_tweets:
    tweet_text = tweet["text"]
    #Pretty print in JSON
    #print(json.dumps(tweet_text, indent = 4, sort_keys = True))
    tweets_BBC.append(tweet_text)
    
    # Run Vader Analysis on each tweet
    results = analyzer.polarity_scores(tweet["text"])
    compound = results["compound"]
    pos = results["pos"]
    neu = results["neu"]
    neg = results["neg"]
    tweets_ago = counter
        
    # Add sentiments for each tweet into an array
    sentiments_BBC.append({"Source Account":target_term,
                           "Tweet Text": tweet["text"],
                        "Date": tweet["created_at"], 
                       "Compound": compound,
                       "Positive": pos,
                       "Negative": neu,
                       "Neutral": neg,
                       "Tweets Ago": counter})
        
    # Add to counter 
    counter = counter + 1
```

```python
# converting the sentiments of BBC News to a dataframe
sentimentsbbc_pd = pd.DataFrame.from_dict(sentiments_BBC)
sentimentsbbc_pd.head()
sentimentsbbc_pd.to_csv("tweet_details.csv")
```

```python
# Creating ScatterPlot
plt.title("Sentiment Analysis of Media Tweets(6/14/2018)")
plt.xlabel("Tweets Ago")
plt.ylabel("Tweet Polarity")
plt.scatter(np.arange(len(sentimentsbbc_pd["Compound"])),sentimentsbbc_pd["Compound"], marker = "o",alpha=.8, color = "cyan",edgecolors="black",label="BBC")
plt.scatter(np.arange(len(sentimentscnn_pd["Compound"])),sentimentscnn_pd["Compound"], marker = "o", alpha=.8,color = "red",edgecolors="black",label="CNN")
plt.scatter(np.arange(len(sentimentscbs_pd["Compound"])),sentimentscbs_pd["Compound"], marker = "o", alpha = 0.8, color = "purple",edgecolors="black",label="CBS")
plt.scatter(np.arange(len(sentimentsfox_pd["Compound"])),sentimentsfox_pd["Compound"], marker = "o", alpha = 0.8, color = "green",edgecolors="black",label="Fox News")
plt.scatter(np.arange(len(sentimentsnytimes_pd["Compound"])),sentimentsnytimes_pd["Compound"], marker = "o", alpha = 0.8, color = "yellow",edgecolors="black",label="NewYork Times")
plt.legend(fontsize = "small", mode="expanded",numpoints=1,loc="best", title = "Media Sources")
plt.grid(True)

#save figure
plt.savefig("sentiment_analysis.png")
plt.show()
```
