```python
# Temperature Analysis
#Write a function called calc_temps that will accept a start date and end date in the format %Y-%m-%d and return the minimum, average, and maximum temperatures for that range of dates.
#Use the calc_temps function to calculate the min, avg, and max

def cal_temps():
    start_date = input("Enter a Start Date:")
    end_date = input("Enter an End Date:")
    format_str = '%Y-%m-%d'
    start_date1 = datetime.strptime(start_date, format_str)
    end_date1 = datetime.strptime(end_date, format_str)
    print(start_date1.date())
    print(end_date1.date())
    next_year = (start_date1.year)+1
    
    
    maximum = session.query(Station.station, Station.name, Measurement.date,Measurement.tobs, func.max(Measurement.tobs)).filter(Measurement.date.between(start_date1, end_date1))
    minimum = session.query(Station.station, Station.name, Measurement.date,Measurement.tobs, func.min(Measurement.tobs)).filter(Measurement.date.between(start_date1, end_date1))
    average = session.query(Station.station, Station.name, Measurement.date,Measurement.tobs, func.avg(Measurement.tobs)).filter(Measurement.date.between(start_date1, end_date1))
    for m in maximum:
        print("Maximum is:", m[4]) 
        
    for mi in minimum:
        print("Minimum is:", mi[4])
    
    for av in average:
        print("Average is:", av[4])
    return
```

```python
# running the function which inputs the start and end date of the vacation and returns the maximum, minimum
# and average temperature for the trip
cal_temps()
```

```python

def cal_map():
    start_date = input("Enter a Start Date:")
    end_date = input("Enter an End Date:")
    format_str = '%Y-%m-%d'
    start_date1 = datetime.strptime(start_date, format_str)
    end_date1 = datetime.strptime(end_date, format_str)
    #print(start_date1.date())
    #print(end_date1.date())
    previous_year = (start_date1.year)-1
    dt = datetime(previous_year, start_date1.month, start_date1.day)
    print(dt)
    str_date = datetime.strftime(dt, format_str)
    print(str_date)
    nstart_date = datetime.strptime(str_date, format_str)
    print(nstart_date)
    
    
    
    maximum = session.query(Station.station, Station.name, Measurement.date,Measurement.tobs, func.max(Measurement.tobs)).filter(Measurement.date.between(nstart_date, start_date1))
    minimum = session.query(Station.station, Station.name, Measurement.date,Measurement.tobs, func.min(Measurement.tobs)).filter(Measurement.date.between(nstart_date, start_date1))
    average = session.query(Station.station, Station.name, Measurement.date,Measurement.tobs, func.avg(Measurement.tobs)).filter(Measurement.date.between(nstart_date, start_date))
       
    for m in maximum:
        max_value = m[4]
        print(max_value)
        print("Maximum for previous year is:", m[4])
    
        
    for mi in minimum:
        min_value = mi[4]
        print("Minimum for previous year is:", mi[4])
   
    yerr = max_value - min_value
    print(yerr)
    means = (max_value - min_value)/2
    
    for av in average:
        avg = av[4]
        print(avg)
        

    # setting up the plot
    fig, ax = plt.subplots()

    ax.bar(means, yerr, label = "Trip Average Temperature")

    plt.savefig("temp_avg.png")
    plt.show()
    
```
