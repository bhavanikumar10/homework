```python
# creating an engine to create an sqlite database file called "hawaii.sqlite"
engine = create_engine("sqlite:///hawaii.sqlite")
```

```python
# creating a connection to the engine
conn = engine.connect()
```

```python
# using declarative_base() to create a ORM for the class "measurement" and "station"
Base = declarative_base()

class Measurement(Base):
    __tablename__ = "measurements"
    id = Column(Integer, primary_key = True)
    date = Column(String)
    prcp = Column(Float)
    station = Column(String)
    tobs = Column(Integer)
    
    def __repr__(self):
        return f"date={self.date}, id={self.id},precipitation = {self.precipiation},station={self.station}, tobs={self.tobs}"

class Station(Base):
    __tablename__ = "stations"
    id = Column(Integer, primary_key=True)
    station = Column(String)
    name = Column(String)
    latitude = Column(Float)
    longtitude = Column(Float)
    elevation = Column(Float)
    
    def __repr__(self):
        return f"station={self.station}, name={self.name}, latitude = {self.latitude}, longtitude = {self.longtitude}, elevation = {self.elevation}"

```

```python
# creating a reference to the created classes, to reflect the tables
Base.metadata.create_all(engine)
metadata = MetaData(bind=engine)
metadata.reflect()
```

```python
# Save the reference to the `measurements` table as a variable called `mtable`. 
# Saving the reference to the stations table as a variable called 'stable'
mtable = sqlalchemy.Table('measurements', metadata, autoload=True)
stable = sqlalchemy.Table('stations', metadata, autoload = True)
```
